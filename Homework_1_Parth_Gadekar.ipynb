## Question 3
Generate an array with 100 rows and 2 cols at random. Plot it by Matplotlib. You should consider the plotting setting in terms of titles, legends, names of the x and y axe, etc.

**Firstly, please set your student id as the seed number. Like np.random.seed(20016329).**

```python
# Step 1: Set the random seed using the student ID
np.random.seed(20029972)

# Step 2: Generate a random array with 100 rows and 2 columns
data = np.random.rand(100, 2)

# Step 3: Scatter plot the data
plt.figure(figsize=(7,6))
plt.scatter(data[:, 0], data[:, 1], color='black', label='Data Points')

# Step 4: Enhance the plot
plt.title("Scatter Plot of Randomly Generated Data from a uniform distribution over [0,1)", fontsize=14)
plt.xlabel("Col 1", fontsize=12)
plt.ylabel("Col 2", fontsize=12)
plt.legend(loc='upper right', fontsize=10)
plt.xticks(np.arange(0, 1.2, 0.1))

# Display the plot
plt.show()
```

## Question 4
For the previous generated data, consider each row of the array as an individual data point, with the two columns representing its distinct features. Identify the data points with the smallest and largest sum of features. Mark them in red and blue in the figure respectively.

Hint: You can compute the sum of each row and store it in an array named "sums". Then find the minimum or maximum index of this array.

```python
# Step 1: Set the random seed using the student ID
np.random.seed(20029972)

# Step 2: Generate a random array with 100 rows and 2 columns
data = np.random.rand(100, 2)

# Step 3: Compute the sum of each row
sums = np.sum(data, axis=1)

# Step 4: Find the index of the smallest and largest sums
min_index = np.argmin(sums)
max_index = np.argmax(sums)

# Step 5: Scatter plot the data
plt.figure(figsize=(7,6))
plt.scatter(data[:, 0], data[:, 1], color='black', label='Data Points')

# Step 6: Highlight the point with the smallest sum in red and the largest in blue
plt.scatter(data[min_index, 0], data[min_index, 1], color='red', label='Smallest Sum')
plt.scatter(data[max_index, 0], data[max_index, 1], color='blue', label='Largest Sum')

# Step 7: Enhance the plot
plt.title("Data Points with Extremes Marked", fontsize=14)
plt.xlabel("Feature 1", fontsize=12)
plt.ylabel("Feature 2", fontsize=12)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3, fontsize=10)
plt.xticks(np.arange(0, 1.1, 0.1))

# Display the plot
plt.show()
```

## Question 5
Add two lines representing the mean of the Feature 1 and the 25% quantile of the Feature 2 seperately. Add the legend of the two lines.

```python
# Step 1: Set the random seed using the student ID
np.random.seed(20029972)

# Step 2: Generate a random array with 100 rows and 2 columns
data = np.random.rand(100, 2)

# Step 3: Compute the sum of each row
sums = np.sum(data, axis=1)

# Step 4: Compute the mean of Feature 1 and the 25% quantile of Feature 2
mean_feature1 = np.mean(data[:, 0])
quantile_feature2 = np.percentile(data[:, 1], 25)

# Step 5: Scatter plot the data
plt.figure(figsize=(7,6))
plt.scatter(data[:, 0], data[:, 1], color='black', label='Data Points')

# Step 6: Plot the mean of Feature 1 (vertical line) and 25% quantile of Feature 2 (horizontal line)
plt.axvline(x=mean_feature1, color='green', linestyle='-', label=f'Mean of Feature 1')
plt.axhline(y=quantile_feature2, color='orange', linestyle='-', label=f'25% Quantile of Feature 2')

# Step 7: Enhance the plot
plt.title("Scatter Plot with Mean and Quantile Lines", fontsize=14)
plt.xlabel("Feature 1", fontsize=12)
plt.ylabel("Feature 2", fontsize=12)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3, fontsize=10)
plt.xticks(np.arange(0, 1.1, 0.1))

# Display the plot
plt.show()
```
